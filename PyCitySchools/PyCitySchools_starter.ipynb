```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data_original= pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data_original, school_data, how="left", on=["school_name", "school_name"])

school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Total number of schools
total_schools = len(school_data_complete["school_name"].unique())

total_schools

# Total Number of students
total_students = len(school_data_complete["student_name"])

total_students

#total Budget
unique_budget = school_data_complete["budget"].unique()
total_budget = unique_budget.sum()
total_budget

# Average Math Score 
avg_math = school_data_complete["math_score"].mean()
avg_math

# Calculate the average reading score
avg_reading = school_data_complete["reading_score"].mean() 
avg_reading

#Calculate the percentage of students with a passing math score (70 or greater)
math_gtr_70 = school_data_complete.loc[school_data_complete["math_score"]>=70]
mathgtr70 = math_gtr_70[["Student ID", "math_score"]]
mathpct_gtr_70 = len(math_gtr_70.value_counts())/total_students
mathpct_gtr_70




#Calculate the percentage of students with a passing reading score (70 or greater)
readScores= school_data_complete.loc[school_data_complete["reading_score"] >=70]
readgtr70 = readScores[["Student ID", "reading_score"]]
readpct_gtr_70 = len(readScores.value_counts())/total_students
readpct_gtr_70


#Calculate the percentage of students who passed math and reading (% Overall Passing)
overall_score_df = pd.merge(readgtr70, mathgtr70, on="Student ID")

scores_gtr_70 = overall_score_df.loc[(overall_score_df["reading_score"]>=70)|
                                     (overall_score_df["math_score"]>=70)]
overallpct_gtr_70 = len(scores_gtr_70)/total_students
overallpct_gtr_70

#Create a dataframe to hold the above results
district_summary_df = pd.DataFrame({"Total Schools": [total_schools],
                              "Total Students": [total_students],
                              "Total Budget":[total_budget],
                              "Avg Math Score": avg_math,
                              "Avg Reading Score": avg_reading,
                              "Passing Math Scores": mathpct_gtr_70,
                              "Passing Reading Scores": readpct_gtr_70,
                              "Passing Math & Reading Scores":overallpct_gtr_70 })
district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#School Name, School Type, Total Students, Total School Budget
school_data_df = school_data
original_summary_df = school_data_df.set_index("school_name")
original_summary_df.sort_values("school_name")
summary_renamed_df=original_summary_df.rename(columns = {"type": "School Type",
                                                        "size": "Total Students",
                                                        "budget": "Total Budget"})


del summary_renamed_df['School ID']
summary_renamed_df

#Per Student Budget
per_student_budget = summary_renamed_df["Total Budget"]/summary_renamed_df["Total Students"]
summary_renamed_df["Budget Per Student"]= per_student_budget
summary_renamed_df.sort_values("school_name")

#avg math and reading score per school
student_data=student_data_original[["school_name","math_score","reading_score"]]
student_data_group = student_data.groupby(["school_name"])
student_data_avg = student_data_group.mean()
student_data_avg

merge_df = pd.merge(summary_renamed_df,student_data_avg, on="school_name")
student_data_new = merge_df.rename(columns = {"math_score":"Avg Math Score",
                          "reading_score":"Avg Reading Score"})
#passing Math Scores count
student_data_math = student_data[["school_name","math_score"]]
student_data_70 = student_data_math.loc[student_data_math["math_score"]>=70]
student_data_70group = student_data_70.groupby(["school_name"])
student_data_gtr70 = student_data_70group.count()
student_data_gtr70 = student_data_gtr70.rename(columns = {"math_score":"% Passing Math"})



#passing reading scores count
student_data_reading = student_data[["school_name","reading_score"]]
student_data_reading_70 = student_data_reading.loc[student_data_reading["reading_score"]>=70]
student_reading_70group = student_data_reading_70.groupby(["school_name"])
student_dataread_gtr70 = student_reading_70group.count()
student_dataread_gtr70 = student_dataread_gtr70.rename(columns = {"reading_score": "% Passing Reading"})


#overall passing count
student_data_m70 = student_data.loc[(student_data["math_score"]>=70)]
student_data_m_r70=student_data_m70.loc[(student_data["reading_score"]>=70)]
student_data_m_r70
student_group_m_r70 = student_data_m_r70.groupby(["school_name"])
student_count_m_r70 =student_group_m_r70.count() 
student_count_m_r70=student_count_m_r70[["math_score"]]
student_count_m_r70 = student_count_m_r70.rename(columns={"math_score":"% Overall Passing"})


student_data_complete = pd.merge(student_data_new,student_data_gtr70, on= "school_name")
student_data_complete = pd.merge(student_data_complete,student_dataread_gtr70, on= "school_name")
student_data_complete = pd.merge(student_data_complete,student_count_m_r70, on= "school_name")

student_data_complete["% Passing Math"]= student_data_complete["% Passing Math"]/student_data_complete["Total Students"]*100
student_data_complete["% Passing Reading"]= student_data_complete["% Passing Reading"]/student_data_complete["Total Students"]*100
student_data_complete["% Overall Passing"]= student_data_complete["% Overall Passing"]/student_data_complete["Total Students"]*100

student_data_complete.sort_values("school_name")
school_summary_data_complete = student_data_complete.sort_values("school_name")
school_summary_data_complete
#student_data_complete.sort_values("school_name")
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
student_math = student_data_original[["school_name","grade","math_score"]]

schools = []
math_9 = []
math_10 = []
math_11 = []
math_12 = []


student_math_sort = student_math.groupby(["school_name"]).mean()
new_student_math_sort=student_math_sort.reset_index()
for school in new_student_math_sort["school_name"]:
    schools.append(school)


student_math9 = student_math.loc[student_math["grade"]=="9th"]
mean_math9=student_math9.groupby(["school_name"]).mean()
for score in mean_math9["math_score"] :
    math_9.append(score)
math_9

student_math10 = student_math.loc[student_math["grade"]=="10th"]
mean_math10=student_math10.groupby(["school_name"]).mean()
for score in mean_math10["math_score"] :
    math_10.append(score)
math_10

student_math11 = student_math.loc[student_math["grade"]=="11th"]
mean_math11=student_math11.groupby(["school_name"]).mean()
for score in mean_math11["math_score"] :
    math_11.append(score)
math_11

student_math12 = student_math.loc[student_math["grade"]=="12th"]
mean_math12=student_math12.groupby(["school_name"]).mean()
for score in mean_math12["math_score"] :
    math_12.append(score)
math_12

by_grade_school = pd.DataFrame({"9th": math_9,
                              "10th": math_10,
                              "11th": math_11,
                              "12th": math_12,},
                              index = schools)

by_grade_school 
```
